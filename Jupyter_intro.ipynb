## Notebook grensesnittet
Forhpentligvis ser ikke en Jupyter notebook helt fremmed ut. Strengt tatt er dette en noe avansert tekstbehandler som ogs lar deg kjre python.

Nr du pnet en tom notebook s den slik ut:

Dette er en **celle**.

**Celler** og **kjerner (kernels)** er to begreper vi m vite hva betyr.
* En **kernel** er 'maskinen' som kjrer i bakgrunnen og behandler koden som skrives i en notebook. Den tolker programsprket vi skriver og returnerer output til notebooken.
* En **celle** er en boks som inneholder tekst, formler og figurer som skal vises i notebooken, eller som inneholder kode som skal kjres av notebookens kernel.

Det er to typer celler, som er beregnet p forskjellig innhold.

## Hurtigtaster
Noen hurtigtaster m vi kjenne til. Om ikke annet fordi vi kan komme borti disse ved uhell.

Det vil alltid vre en **aktiv celle** i en notebook. Du har kanskje sett at denne har **grnn** kantlinje nr du redigerer cella, men at kantlinjen blir **bl** nr du kjrer cella.
> Grnn kantlinje -- cella er i redigeringsmodus  
> Bl kantlinje -- cella er i kommandomodus

Redigeringsmodus betyr at du kan skrive inn tekst eller kode i cella. Kodeceller settes i redigeringsmodus ved  trykke i kodefeltet. Markdown celler settes i redigeringsmodus ved  dobbeltklikke p cella.

Kommandomodus lar oss enkelt flytte rundt p celler. Vi trykker i margen til venstre for en celle for  velge den uten  samtidig g inn i redigeringsmodus.

Her er en oversikt over noen av de vanligste og mest brukbare hurtigtastene i Jupyter.

* Skift mellom redigering- og kommandomodus med `Esc` og `Enter`.
I kommandomodus:
* Naviger mellom celler med piltastene `Opp` og `Ned`.
* Trykk `A` eller `B` for  legge til en celle over elle runder den aktive cellen.
* `M` gjr om aktiv celle til en markdown celle.
* `Y` gjr om aktiv celle til en kodecelle.
* `D + D` (`D` to ganger) **sletter** aktiv celle.
* `Z`angrer sletting av en celle.
* Hold inne `Shift` og naviger `Opp` eller `Ned` for  markere flere celler.
* Nr flere celler er valgt samtidig kan de sls sammen med `M`.\
I redigeringsmodus:
* `Ctrl + Shift + -` deler cella i to ved markren.

Pass spesielt p disse hvis du skal skrive noe i notebooken og ved et uhell ikke er i redigeringsmodus!

# Kernels
Bak enhver notebook kjrer en kernel. Nr vi kjrer en celle blir koden utfrt av kernelen. Hvis koden produserer et output sendes dette tilbake til cellen for  vises.

Hvis en kodeblokk oppretter variabler eller data lagres dette i kernelens minne. Det er viktig  vite at kernelens minne eller tilstand bevares over tid og ogs mellom celler.

For eksempel, hvis vi importerer et bibliotek eller deklarerer en variabel i en celle gjelder dette ogs for alle andre celler som kjres i ettertid.

````python
import numpy as np
import matplotlib.pyplot as plt
def f(x):
    return 2*x**2 + 3
````

N er biblioteket `numpy` importert med forkortelsen `np` og vi har definert funksjonen $f(x) = 2x^2+3$. Vi har ogs importert `pyplot`, som lar oss plotte funksjoner.

Vi kan n kalle funksjonen, og ogs bruke funksjoner fra numpy overalt ellers i notebooken *s lenge vi har kjrt cella over*.

````python
x = np.linspace(0,12,50)
y = f(x)
plt.plot(x,y)
plt.grid()
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.show()
````

I cellen over definerer vi en array `x` (liste) ved bruk av funksjonen `np.linspace(0,12,50)`. Denne funksjonen kommer fra `numpy` og oppretter en en liste med frste verdi 0 og siste verdi 12, og til sammen 50 verdier spredt jevnt over dette intervallet.
Deretter definerer vi en array `y` som anvender funksjonen $f(x) = 2x^2 +3$ p lista `x`, alts den finner funksjonsverdien $y = f(x)$ for intervallet $x\in[0,12]$.
Hensikten med  bruke s mange som 50 verdier er for at grafen skal f fin opplsning.
(Prv for eksempel  endre koden til `np.linspace(0,12,5)`).

N skal vi restarte kernel for  se hva som kan g galt nr kode er fordelt over flere celler. Restart kernel ved  g inn i menyen **Kernel** og velg **Restart & Clear Output**.

Etterp prver du  kjre cellen over.

N fr vi en feilmelding:


Du vil se at det n str `In[1]` ved siden av cella vi kjrte, men ved alle andre kodeceller str det `In[]`. Det betyr at disse cellene ikke enda har kjrt. Dermed vet ikke enda kernel at biblioteket `numpy` har blitt lastet med forkortelse `np`, og vi fr denne feilmeldingen.

For  unng slikt m vi passe p at cellene kjres i korrekt rekkeflge.

Vi kan i denne omgang lse problemet med **Run All** slik at alle celler kjres i rekkeflge fra verst til nederst. Gjr dette n.\
Husk at dette kan ta litt tid! Du kan se det vil st `In[*]` ved siden av celler som venter p  bli kjrt, eller er i ferd med  kjres.

Dette m vi ogs tenke p hvis vi f. eks vil endre $f(x)$ til $f(x) = \sqrt{x}$. Da m vi *frst* endre definisjonen av `f(x)` i den verste cella, deretter kjre den verste cella slik at dette oppdateres i kernel, og til sist kjre den nederste cella p nytt slik at `y` og figuren oppdateres.

For det meste vil flyten i en notebook vre ovenfra og nedover, men det er vanlig  g tilbake for  gjre endringer. I s fall m vi holde styr p hvilken rekkeflge kodecellene kjres i. Input indikatoren, slik som `In[4]`, hjelper oss  finne frem til eventuelle feil som oppstr p grunn av kjringsrekkeflgen til cellene.

De tre restartfunksjonene fra Kernel menyen fungerer slik:
* Restart: Restarter kernelen. Da slettes kernelens minne, og variabler osv som tidligere ble deklarert er n borte.
* Restart & Clear Output: Gjr det samme, men n fjernes ogs all output som vises under kodecellene.
* Restart & Run All: Samme som over, men etterp kjres alle kodecellene i rekkeflge.

Hvis kernel noen ganger setter seg fast i en beregnig kan den avbrytes med **Interrupt** fra kernelmenyen, eller ved  trykke **stoppknappen**. Dette er en viktig funksjon, med hurtigtast `I + I (to ganger)` fra kommandomodus.

For eksempel kan det vre ndvendig  avbryte en celle som venter p input.

````python
a = int(input('Skriv inn et tall: '))
b = int(input('Skriv in enda et tall: '))

print('Summen av tallene er', a+b)
````

Frst en forklaring av hva denne koden gjr.

Funksjonen `input` skaper et prompt hvor brukeren m skrive inn en verdi. Denne verdien lagres som en tekststreng (string) i minnet.
Funksjonen `int` gjr om tekstrengen til heltall, s lenge inntastet verdi faktisk var et heltall. Du vil f en feilmelding her om du skriver inn noe tull.

Deretter skriver funksjonen `print` ut 'Summen av tallene er' og verdien `a + b`. Operatoren `+` er annerledes for heltall enn for string. Prv  fjern `int` i koden over, s ser du hva jeg mener.

Nr en celle venter p input settes kernel p **hold**. Det vil si at ingen andre celler kan kjres i mellomtiden. Hvis du prver  kjre en annen celle vil den settes p vent. Vi m derfor gi kernel den input den forventer nr vi frst har startet. Om vi skulle nske  komme oss ut av dette uten  gi forventet input m vi bruke **Interrupt**. Det kan vre relevant hvis vi for eksempel ikke vet hva vi driver med, og har misforsttt hvilken input kernel venter p.

Andre viktige situasjoner hvor vi m vre kjent med interrupt er for eksempel ved bruk av **lkker**.

Kjr koden under og se hvordan det foregr.

````python
n = 0
while(n < 12):
    n = n + 1
    print(n)
````

I denne kodecella har vi brukt funksjonen `while`. Den fungerer slik at den kjrer kode hvis en logisk funksjon har verdien `True`.

I vrt tilfelle sjekker den om verdien til variabelen `n` er mindre enn 12, og hvis dette er sant blir `n` satt til `n + 1`.

Etterp endrer du koden slik at det heller trekkes fra 1 hvis `n` er mindre enn 12:  
`n = n - 1` erstatter `n = n + 1`.
Da vil aldri den logiske betingelsen `n < 12` bli usann, og while-lkka vil aldri slutte.

Prv n  kjr cella p nytt.  
**Obs!** Det kommer til  g galt, s trykk interrupt etter du har kjrt cella.

Vi ser at dette ikke gikk srlig bra etter  ha kjrt koden. Min maskin jobbet frem til `n = -34231` fr jeg rakk  trykke interrupt. Hvis en slik uendelig lkke aldri sluttes vil programmet heller aldri g videre. I dette tilfellet ville nok etter hvert tallet `n` blitt s stort at kernel gr tom for minne. Da vil den stenges ned av seg selv, eller eventuelt av operativsystemet.

I tilfeller med lkker som gr i loop, eller lkker som tar svrt lang tid  kjre, kan vi vre ndt til  bruke **interrupt** for s  fikse feil i koden eller kanskje gjre den smartere slik at den kjrer raskere.
