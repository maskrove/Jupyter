## Notebook grensesnittet
Forhåpentligvis ser ikke en Jupyter notebook helt fremmed ut. Strengt tatt er dette en noe avansert tekstbehandler som også lar deg kjøre python.

Når du åpnet en tom notebook så den slik ut:


Menyene og verktøylinja utforsker vi senere. Først må vi ta hensyn til denne mystiske boksen innrammet i grønnt på bildet.

Dette er en **celle**.

**Celler** og **kjerner (kernels)** er to begreper vi må vite hva betyr.
* En **kernel** er 'maskinen' som kjører i bakgrunnen og behandler koden som skrives i en notebook. Den tolker programspråket vi skriver og returnerer output til notebooken.
* En **celle** er en boks som inneholder tekst, formler og figurer som skal vises i notebooken, eller som inneholder kode som skal kjøres av notebookens kernel.

Det er to typer celler, som er beregnet på forskjellig innhold.

## Celler
De to typene celler er:
* Kode-celler, som inneholder kode som skal kjøres av kernel. Når en kode-celle kjøres vises output rett under cellen.
* Markdown-celler, som inneholder tekst formattert etter [Markdown](https://www.markdownguide.org/basic-syntax/) systemet. Det er mange formatteringsmuligheter i Markdown, som for eksempel lister, tabeller, figurer og formler. All tekst og alle bilder og lister så langt i denne veiledningen er skrevet i slike markdown-celler. Markdown celler omgjøres til formatert tekst når de kjøres.

Begge typer celler kjøres ved å trykke **Run** på verktøylinja, eller med hurtigtasten `Ctrl + Enter`.

Merk at dette kjører kun cella du har markert. Hvis du vil kjøre **alle cellene** i hele notebooken må du inn i menyen **Cell** og deretter velge **Run all**. Da kjøres cellene i rekkefølge fra øverst til nederst.

## Hurtigtaster
Noen hurtigtaster må vi kjenne til. Om ikke annet fordi vi kan komme borti disse ved uhell.

Det vil alltid være en **aktiv celle** i en notebook. Du har kanskje sett at denne har **grønn** kantlinje når du redigerer cella, men at kantlinjen blir **blå** når du kjører cella.
> Grønn kantlinje -- cella er i redigeringsmodus  
> Blå kantlinje -- cella er i kommandomodus

Redigeringsmodus betyr at du kan skrive inn tekst eller kode i cella. Kodeceller settes i redigeringsmodus ved å trykke i kodefeltet. Markdown celler settes i redigeringsmodus ved å dobbeltklikke på cella.

Kommandomodus lar oss enkelt flytte rundt på celler. Vi trykker i margen til venstre for en celle for å velge den uten å samtidig gå inn i redigeringsmodus.

Her er en oversikt over noen av de vanligste og mest brukbare hurtigtastene i Jupyter.

* Skift mellom redigering- og kommandomodus med `Esc` og `Enter`.
I kommandomodus:
* Naviger mellom celler med piltastene `Opp` og `Ned`.
* Trykk `A` eller `B` for å legge til en celle over elle runder den aktive cellen.
* `M` gjør om aktiv celle til en markdown celle.
* `Y` gjør om aktiv celle til en kodecelle.
* `D + D` (`D` to ganger) **sletter** aktiv celle.
* `Z`angrer sletting av en celle.
* Hold inne `Shift` og naviger `Opp` eller `Ned` for å markere flere celler.
* Når flere celler er valgt samtidig kan de slås sammen med `M`.\
I redigeringsmodus:
* `Ctrl + Shift + -` deler cella i to ved markøren.

Pass spesielt på disse hvis du skal skrive noe i notebooken og ved et uhell ikke er i redigeringsmodus!

## Markdown
Hele denne introduksjonen er skrevet i en Jupyter notebook. Det betyr at all forklarende tekst, bilder og øvrig formatering er skrevet i Markdown.

Du kan dobbeltklikke i en vilkårlig celle i dette dokumentet som inneholder overskrifter, lister, linker, bilder eller andre typer formatering for å se hvordan dette ble gjort i markdown.

Videre utdyping finner du i [oversikten](https://www.markdownguide.org/basic-syntax/) over grunnleggende Markdown syntaks.

For at det skal være enklest mulig å finne frem til en bestemt type formattering gjentas de vanligste typene her.

# Dette er en nivå 1 overskrift
## Dette er en nivå 2 overskrift (--- under gir tynn understreking)

Dette er ren tekst. Ren tekst krever ingen spesiell formatering, som er til fordel for våre elever. Vi kan fremheve skrift enten som *kursiv* og _kursiv_ eller som **fet** og __fet__ skrift.

Hvis vi vil bruke spesielle symboler som normalt brukes i formatering kan vi gjøre dette med bruk av \ foran symbolet.
Eksempelvis, \* \$ \# osv.

Avsnitt må deles opp med en tom linje.

Vi kan tvinge en linje til å slutte  
ved å skrive to mellomrom der vi vil ha linjeskift.

* Hvert punkt på lista startes med en asterisk \*
    * Vi kan lage underpunkter i en liste ved å indentere punktet med `Tab`


1. Lister kan også være nummerert.
2. I så fall bruker vi tall og punktum for å lage nummererte punkter.

For å fremheve kode underveis i teksten benyttes enkel bakoverlent fjutt (backtick) før og etter `foo()`, og for å fremheve en hel kodeblokk brukes trippelfjutt:

# Kernels
Bak enhver notebook kjører en kernel. Når vi kjører en celle blir koden utført av kernelen. Hvis koden produserer et output sendes dette tilbake til cellen for å vises.

Hvis en kodeblokk oppretter variabler eller data lagres dette i kernelens minne. Det er viktig å vite at kernelens minne eller tilstand bevares over tid og også mellom celler.

For eksempel, hvis vi importerer et bibliotek eller deklarerer en variabel i en celle gjelder dette også for alle andre celler som kjøres i ettertid.

````python
import numpy as np
import matplotlib.pyplot as plt
def f(x):
    return 2*x**2 + 3
````

Nå er biblioteket `numpy` importert med forkortelsen `np` og vi har definert funksjonen $f(x) = 2x^2+3$. Vi har også importert `pyplot`, som lar oss plotte funksjoner.

Vi kan nå kalle funksjonen, og også bruke funksjoner fra numpy overalt ellers i notebooken *så lenge vi har kjørt cella over*.

````python
x = np.linspace(0,12,50)
y = f(x)
plt.plot(x,y)
plt.grid()
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.show()
````

I cellen over definerer vi en array `x` (liste) ved bruk av funksjonen `np.linspace(0,12,50)`. Denne funksjonen kommer fra `numpy` og oppretter en en liste med første verdi 0 og siste verdi 12, og til sammen 50 verdier spredt jevnt over dette intervallet.
Deretter definerer vi en array `y` som anvender funksjonen $f(x) = 2x^2 +3$ på lista `x`, altså den finner funksjonsverdien $y = f(x)$ for intervallet $x\in[0,12]$.
Hensikten med å bruke så mange som 50 verdier er for at grafen skal få fin oppløsning.
(Prøv for eksempel å endre koden til `np.linspace(0,12,5)`).

Nå skal vi restarte kernel for å se hva som kan gå galt når kode er fordelt over flere celler. Restart kernel ved å gå inn i menyen **Kernel** og velg **Restart & Clear Output**.

Etterpå prøver du å kjøre cellen over.

Nå får vi en feilmelding:


Du vil se at det nå står `In[1]` ved siden av cella vi kjørte, men ved alle andre kodeceller står det `In[]`. Det betyr at disse cellene ikke enda har kjørt. Dermed vet ikke enda kernel at biblioteket `numpy` har blitt lastet med forkortelse `np`, og vi får denne feilmeldingen.

For å unngå slikt må vi passe på at cellene kjøres i korrekt rekkefølge.

Vi kan i denne omgang løse problemet med **Run All** slik at alle celler kjøres i rekkefølge fra øverst til nederst. Gjør dette nå.\
Husk at dette kan ta litt tid! Du kan se det vil stå `In[*]` ved siden av celler som venter på å bli kjørt, eller er i ferd med å kjøres.

Dette må vi også tenke på hvis vi f. eks vil endre $f(x)$ til $f(x) = \sqrt{x}$. Da må vi *først* endre definisjonen av `f(x)` i den øverste cella, deretter kjøre den øverste cella slik at dette oppdateres i kernel, og til sist kjøre den nederste cella på nytt slik at `y` og figuren oppdateres.

For det meste vil flyten i en notebook være ovenfra og nedover, men det er vanlig å gå tilbake for å gjøre endringer. I så fall må vi holde styr på hvilken rekkefølge kodecellene kjøres i. Input indikatoren, slik som `In[4]`, hjelper oss å finne frem til eventuelle feil som oppstår på grunn av kjøringsrekkefølgen til cellene.

De tre restartfunksjonene fra Kernel menyen fungerer slik:
* Restart: Restarter kernelen. Da slettes kernelens minne, og variabler osv som tidligere ble deklarert er nå borte.
* Restart & Clear Output: Gjør det samme, men nå fjernes også all output som vises under kodecellene.
* Restart & Run All: Samme som over, men etterpå kjøres alle kodecellene i rekkefølge.

Hvis kernel noen ganger setter seg fast i en beregnig kan den avbrytes med **Interrupt** fra kernelmenyen, eller ved å trykke **stoppknappen**. Dette er en viktig funksjon, med hurtigtast `I + I (to ganger)` fra kommandomodus.

For eksempel kan det være nødvendig å avbryte en celle som venter på input.

````python
a = int(input('Skriv inn et tall: '))
b = int(input('Skriv in enda et tall: '))

print('Summen av tallene er', a+b)
````

Først en forklaring av hva denne koden gjør.

Funksjonen `input` skaper et prompt hvor brukeren må skrive inn en verdi. Denne verdien lagres som en tekststreng (string) i minnet.
Funksjonen `int` gjør om tekstrengen til heltall, så lenge inntastet verdi faktisk var et heltall. Du vil få en feilmelding her om du skriver inn noe tull.

Deretter skriver funksjonen `print` ut 'Summen av tallene er' og verdien `a + b`. Operatoren `+` er annerledes for heltall enn for string. Prøv å fjern `int` i koden over, så ser du hva jeg mener.

Når en celle venter på input settes kernel på **hold**. Det vil si at ingen andre celler kan kjøres i mellomtiden. Hvis du prøver å kjøre en annen celle vil den settes på vent. Vi må derfor gi kernel den input den forventer når vi først har startet. Om vi skulle ønske å komme oss ut av dette uten å gi forventet input må vi bruke **Interrupt**. Det kan være relevant hvis vi for eksempel ikke vet hva vi driver med, og har misforstått hvilken input kernel venter på.

Andre viktige situasjoner hvor vi må være kjent med interrupt er for eksempel ved bruk av **løkker**.

Kjør koden under og se hvordan det foregår.

````python
n = 0
while(n < 12):
    n = n + 1
    print(n)
````

I denne kodecella har vi brukt funksjonen `while`. Den fungerer slik at den kjører kode hvis en logisk funksjon har verdien `True`.

I vårt tilfelle sjekker den om verdien til variabelen `n` er mindre enn 12, og hvis dette er sant blir `n` satt til `n + 1`.

Etterpå endrer du koden slik at det heller trekkes fra 1 hvis `n` er mindre enn 12:  
`n = n - 1` erstatter `n = n + 1`.
Da vil aldri den logiske betingelsen `n < 12` bli usann, og while-løkka vil aldri slutte.

Prøv nå å kjør cella på nytt.  
**Obs!** Det kommer til å gå galt, så trykk interrupt etter du har kjørt cella.

Vi ser at dette ikke gikk særlig bra etter å ha kjørt koden. Min maskin jobbet frem til `n = -34231` før jeg rakk å trykke interrupt. Hvis en slik uendelig løkke aldri sluttes vil programmet heller aldri gå videre. I dette tilfellet ville nok etter hvert tallet `n` blitt så stort at kernel går tom for minne. Da vil den stenges ned av seg selv, eller eventuelt av operativsystemet.

I tilfeller med løkker som går i loop, eller løkker som tar svært lang tid å kjøre, kan vi være nødt til å bruke **interrupt** for så å fikse feil i koden eller kanskje gjøre den smartere slik at den kjører raskere.
