Forhpentligvis ser ikke en Jupyter notebook helt fremmed ut. Strengt tatt er dette en noe avansert tekstbehandler som ogs lar deg kjre python.

Dette er en **celle**.

**Celler** og **kjerner (kernels)** er to begreper vi m vite hva betyr.
* En **kernel** er 'maskinen' som kjrer i bakgrunnen og behandler koden som skrives i en notebook. Den tolker programsprket vi skriver og returnerer output til notebooken.
* En **celle** er en boks som inneholder tekst, formler og figurer som skal vises i notebooken, eller som inneholder kode som skal kjres av notebookens kernel.

Det er to typer celler, som er beregnet p forskjellig innhold.

## Celler
De to typene celler er:
* Kode-celler, som inneholder kode som skal kjres av kernel. Nr en kode-celle kjres vises output rett under cellen.
* Markdown-celler, som inneholder tekst formattert etter [Markdown](https://www.markdownguide.org/basic-syntax/) systemet. Det er mange formatteringsmuligheter i Markdown, som for eksempel lister, tabeller, figurer og formler. All tekst og alle bilder og lister s langt i denne veiledningen er skrevet i slike markdown-celler. Markdown celler omgjres til formatert tekst nr de kjres.

Begge typer celler kjres ved  trykke **Run** p verktylinja, eller med hurtigtasten `Ctrl + Enter`.

Merk at dette kjrer kun cella du har markert. Hvis du vil kjre **alle cellene** i hele notebooken m du inn i menyen **Cell** og deretter velge **Run all**. Da kjres cellene i rekkeflge fra verst til nederst.

## Hurtigtaster
Noen hurtigtaster m vi kjenne til. Om ikke annet fordi det er lett  komme borti disse ved uhell.

Det vil alltid vre en **aktiv celle** i en notebook. Du har kanskje sett at denne har **grnn** kantlinje nr du redigerer cella, men at kantlinjen blir **bl** nr du kjrer cella.
> Grnn kantlinje -- cella er i redigeringsmodus\
> Bl kantlinje -- cella er i kommandomodus

Redigeringsmodus betyr at du kan skrive inn tekst eller kode i cella. Kodeceller settes i redigeringsmodus ved  trykke i kodefeltet. Markdown celler settes i redigeringsmodus ved  dobbeltklikke p cella.
En celle vil forbli i redigeringsmodus frem til den kjres.

Kommandomodus lar oss enkelt flytte rundt p celler. Vi trykker i margen til venstre for en celle for  velge den uten  samtidig g inn i redigeringsmodus.

Her er en oversikt over noen av de vanligste og mest brukbare hurtigtastene i Jupyter.

* Skift mellom redigering- og kommandomodus med `Esc` og `Enter`.
I kommandomodus:
* Naviger mellom celler med piltastene `Opp` og `Ned`.
* Trykk `A` eller `B` for  legge til en celle over elle runder den aktive cellen.
* `M` gjr om aktiv celle til en markdown celle.
* `Y` gjr om aktiv celle til en kodecelle.
* `D + D` (`D` to ganger) **sletter** aktiv celle.
* `Z`angrer sletting av en celle.
* Hold inne `Shift` og naviger `Opp` eller `Ned` for  markere flere celler.
* Nr flere celler er valgt samtidig kan de sls sammen med `M`.

Pass spesielt p disse hvis du skal skrive noe i notebooken og ved et uhell ikke er i redigeringsmodus!
