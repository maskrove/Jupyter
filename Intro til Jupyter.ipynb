# Velkommen til Jupyter
Jupyter notebook er et verkty som lar deg skrive og kjre kode, og som samtidig integrerer output, tekst, formler og figurer i ett enkelt dokument.

Jupyter sttter flere programmeringssprk, deriblant python, og kan kjres enten lokalt p din egen maskin eller p en server. Det som er likt i begge tilfeller er at du fr tilgang til Jupyter notebook via nettleseren. Det gjr det svrt enkelt  bruke etter frstegangs installasjon.

Ved  flge denne introduksjonen vil du forhpentligvis bli fortrolig med grunnleggende bruk av Jupyter, og du vil f litt trening i bruk av Python i matematikk!

## Installasjon
Den enkleste mten  f en fungerende installasjon av Jupyter Notebook er via Anaconda. [Last ned](https://www.anaconda.com/products/individual) og installer den siste utgaven av Anaconda for ditt operativsystem fr du starter.

Installasjonen av Anaconda inneholder bde Python 3 og en hel rekke nyttige verkty som kan brukes for databehandling, blant annet Jupyter. Det som er minst like viktig er at Anaconda ogs inneholder de strste og mest brukte /bibliotekene/ til Python, som for eksempel [NumPy](http://www.numpy.org/), [pandas](https://pandas.pydata.org/), [Matplotlib](https://matplotlib.org/) og mange fler.

Nr du har installert Anaconda p din egen maskin m du pne dette dokumentet i Jupyter. Da vil du f full tilgang til denne introduksjonen.
