# Velkommen til Jupyter
Jupyter notebook er et verktøy som lar deg skrive og kjøre kode, og som samtidig integrerer output, tekst, formler og figurer i ett enkelt dokument.

Jupyter støtter flere programmeringsspråk, deriblant python, og kan kjøres enten lokalt på din egen maskin eller på en server. Det som er likt i begge tilfeller er at du får tilgang til Jupyter notebook via nettleseren. Det gjør det svært enkelt å bruke etter førstegangs installasjon.

Ved å følge denne introduksjonen vil du forhåpentligvis bli fortrolig med grunnleggende bruk av Jupyter, og du vil få litt trening i bruk av Python i matematikk!

## Installasjon
Den enkleste måten å få en fungerende installasjon av Jupyter Notebook er via Anaconda. [Last ned](https://www.anaconda.com/products/individual) og installer den siste utgaven av Anaconda for ditt operativsystem før du starter.

Installasjonen av Anaconda inneholder både Python 3 og en hel rekke nyttige verktøy som kan brukes for databehandling, blant annet Jupyter. Det som er minst like viktig er at Anaconda også inneholder de største og mest brukte /bibliotekene/ til Python, som for eksempel [NumPy](http://www.numpy.org/), [pandas](https://pandas.pydata.org/), [Matplotlib](https://matplotlib.org/) og mange fler.

Når du har installert Anaconda på din egen maskin må du åpne dette dokumentet i Jupyter. Da vil du få full tilgang til denne introduksjonen.

## Kjøre Jupyter

For å åpne en Jupyter notebook, slik som denne du nå skal ha lastet ned, naviger til mappen du lagret fila og klikk for å åpne.
Hvis du vil lage en ny Jupyter Notebook velger du "New" øverst til venstre og deretter "Python 3".
